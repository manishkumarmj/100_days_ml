 # Pairplot
 Pairplots are used to plot multiple paimise bivariate distributions in a dataset. This creates a matrix of axes and shows the
relationship for each pair of columns in a DataFrame. By default, it also draws the univariate distribution of each variable on
the diagonal Axes.


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="Species")
```
