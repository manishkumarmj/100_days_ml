Heat maps 

A heatmaps is basically a chart which is drawn between the categeroical/categeroical generally  

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python



flights =sns.load_dataset('flights')
flights.head()

```

```python
# Alternative using pivot (no need for aggfunc since no duplicates)
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Monthly Passengers by Year')
plt.tight_layout()
plt.show()
x = flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar= False, linewidths= 0.7, annot=True, fmt='d')
# annot = annotation call the value inside the box 
# d = decimal 
```

```python
sns.heatmap(x, cbar= False, linewidths= 0.7, annot=True, fmt='d',cmap='jet')
```
