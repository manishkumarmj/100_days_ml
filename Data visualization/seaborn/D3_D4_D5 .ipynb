```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
tips = sns.load_dataset('tips')
tips.head()

```

```python
tips.head()
```

```python
sns.boxplot(tips['tip'])
```

```python

# we are catplot because we are using figure level function 

sns.catplot(x = 'day', y = 'total_bill', kind= 'box', data= tips   )
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind= 'box',hue= 'sex', data= tips   )
```

# violion Plots 

a. violin plots are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density
plot.
entire plot service  

b. The unquestionable advantage of the violin plot over the box plot is that aside from showing the
abovementioned statistics it also shows the entire distribution of the data.

c. Violin plots are often used to compare the distribution of a given variable across some categories.

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x = 'day', y= 'total_bill',kind= 'violin', hue= 'sex',data=tips )
#sns.catplot(x = 'day', y ='total_bill', kind= 'box',hue= 'sex', data= tips   )
```

```python
sns.catplot(x = 'day', y= 'total_bill',kind= 'violin', hue= 'sex',data=tips,split = True )
```

# Bar PLots 

study on black point 

```python
sns.catplot(x= 'smoker', y= 'total_bill',kind='bar',data=tips)
```

```python
sns.catplot(x= 'smoker', y= 'total_bill',kind='bar',hue='sex',data=tips)
```

use estimeter to calculate the vlaue of the mean mode median 

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', data=tips, estimator=np.median)
```

# count plot   
count plot does the same on catergeroical data 

```python
sns.catplot(x='sex',kind='count', data = tips, hue= 'smoker'  )
```
