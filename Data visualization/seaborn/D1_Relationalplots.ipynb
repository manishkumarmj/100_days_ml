```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
plt.style.use( 'fivethirtyeight' )
```

```python
data = sns.load_dataset('tips') # built in data set 
data.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data=data) # data = provide that variable which you have stored your data init 
# kind  --> define your plot * which type of plot you want to do *
```

# The Hue parameter

--> The hue parameter is used to color the points in a scatterplot. This allows us to create different groups and then color them differently.

```python
sns.relplot(x  = 'total_bill', y = 'tip',hue='smoker',data=data)
```

```python
sns.relplot(x  = 'total_bill', y = 'tip',hue='smoker',style='sex',data=data)
```

```python
sns.relplot(x  = 'total_bill', y = 'tip',hue='smoker',style='sex',size='size',data=data)
```

```python
sns.relplot(x  = 'total_bill', y = 'tip',hue='smoker',style='sex',size='size',sizes=(15,200),data=data)
```
