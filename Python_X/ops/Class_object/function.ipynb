```python
class Fraction:
    # parameterised constructor :- u will pass you items
    def __init__ (self,x,y): # need x and y input 
        self.num = x 
        self.den = y
    def __str__(self): # magic method, print ke andar code dalo str code excute hoga
        return '{}/{}'.format(self.num,self.den)
    def __add__(self,other): # fir1, fr2
        new_num =self.num*other.den + other.num*self.den
        new_den = self.den*other.den ### new_num = 2 * 3 + 1 * 3 # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3  ### new_den = 3 * 3 #### fr1.num*fr2.den + fr2.num*fr1.den # 2/3 + 1/3  # 2/3 + 1/3  # 2/3 + 1/3 
        return '{}/{}'.format(new_num, new_den)
    
    def __sub__(self,other): # fir1, fr2
        new_num =self.num*other.den - other.num*self.den
        new_den = self.den*other.den 
        return '{}/{}'.format(new_num, new_den)
    def __mul__(self,other): # fir1, fr2
        new_num =self.num* other.num
        new_den = self.den*other.den 
        return '{}/{}'.format(new_num, new_den)
    def __truediv__(self,other): # fir1, fr2
        new_num =self.num* other.den 
        new_den = self.den*other.num 
        return '{}/{}'.format(new_num, new_den)
    def __c2deci__(self): # fir1, fr2
        new_deci =self.num/ self.den 
        return new_deci
```

```python
fr1 = Fraction(2,3)
fr2 = Fraction(1,3) # object creation using constructor function
print ("fr1",fr1)
print("fr2",fr2)
print("fr1+fr2",fr1+fr2)
print("fr1-fr2",fr1-fr2)
print("fr1*fr2",fr1*fr2)
print("fr1/fr2",fr1/fr2)
print("fr1 to deci",fr1.__c2deci__())
```
