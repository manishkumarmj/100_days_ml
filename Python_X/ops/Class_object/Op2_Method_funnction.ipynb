```python
class Atm:
    # constructor
    def __init__(self): # this is the constructor  # if u will create an function / variable  then u can use it as a constructor
        print(id(self))
        self.pin =''
        self.balance=0
        #self.menu() # calling menu 
    def menu(self):
        user_input = input("""
              How can i help you 
              1. Press 1 to create a pin
              2. Press 2 to change pin 
              3. Press 3 to check your balance
              4. Press 4 to withdraw
              5. send  money to swiss account modi ji dream work of 15 00 000
              """)
    
        if user_input =='1':
            #Create a pin
            self.CreatePin()
        elif user_input =="2":
            self.changePin()
        # Change pin
        elif user_input =="3":
        #Check your balance
            self.checkBalance()
            
        elif user_input == "4":
        # withdraw money
            self.withdrawMoney()
        else :
            
            exit()
    def CreatePin(self):
        user_pin = input("enter you pin")
        self.pin = user_pin

        user_balance= int(input("enter the balance "))
        self.balance=   user_balance 
        
        print("pin created carefully")
        self.menu()

    def changePin(self):
        oldPin=(input("Enter you old pin"))
        if oldPin== self.pin:
            # change to new pin
            user_pin = input("enter you pin")
            self.pin = user_pin 
            print('pin changed successfull')
            self.menu()
        else:
            print("nai karne de skata re baba")
            self.menu()
    def checkBalance(self):
        print("Your current balance --> ", self.balance) 
        self.menu()
    def withdrawMoney(self):
        withDraw = int(input("Enter the amount"))
        if(withDraw< self.balance):
            print("Your amount is withdrawed ", withDraw)
            self.balance = self.balance-withDraw
            print("Your current balance --> ", self.balance)
            self.menu()
        else:
            print("Ye garib iske maa ke chut")
            self.menu()


## object
```
