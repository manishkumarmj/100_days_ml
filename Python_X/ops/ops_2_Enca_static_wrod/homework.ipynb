```python
class linecl:
    def __init__(self,A,B,C):
        self.a=A;
        self.b=B;
        self.c=C;
    def Point_of_Intersection(self,other ):
        # X
        upper_factorx =(self.b*other.c)-(self.c*other.b)
        lower_factorx= (self.a*other.b)-(self.b*other.a) 
        x= (float)(upper_factorx/lower_factorx)

        # Y
        upper_factoryy=(self.c*other.a)-(self.a*other.c)
        lower_factoryy= (self.a*other.b)-(self.b*other.a)
        y= (float)(upper_factoryy/lower_factoryy)
        return x,y
    # actually don't need 
    def __str__(self):
        return f"{self.a}x + {self.b}y + {self.c} = 0" 
```

```python
# Create two lines
line1 = linecl(2, 3, -5)  # 2x + 3y - 5 = 0
line2 = linecl(1, -1, 0)  # x - y = 0



# Find intersection point
intersection = line1.Point_of_Intersection(line2)
# actually needed 
print(intersection)  # Output: (1.0, 1.0)
# not needed 
# Print line equation
print(line1)  # Output: 2x + 3y + -5 = 0
```
