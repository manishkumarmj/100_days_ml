```python
#sns.displot(titanic['age'],bins =10, kde =False)
plt.hist(df['Age'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)

# bin size will decide the number of portion 


'''
(max- min )/ bin = you bin size and partion 

(80-0)/2  = 40 ka 2 bin banna 

'''
```

```python
plt.hist(df['Age'].dropna(),bins=15, color='skyblue', edgecolor='black', alpha=0.7)
# here dropna() will drop those values whoes not filled
```

```python
# Create histogram with black KDE
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=15, color='skyblue', edgecolor='black', alpha=0.7, kde=True, )

# Add labels and title
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Age Distribution with Black KDE', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv(r'A:\AIML_DEV_FOLDER\ML_X\sefl_Learn\Day_20\train.csv')

# Create a horizontal boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(y=df['Age'].dropna(), color='lightgreen', orient='h')

# Add titles and labels
plt.title('Age Distribution Boxplot', fontsize=14, fontweight='bold')
plt.ylabel('Age', fontsize=12)
plt.xlabel('Value', fontsize=12)

# Remove unnecessary spines
sns.despine(trim=True)

plt.tight_layout()
plt.show()
```
