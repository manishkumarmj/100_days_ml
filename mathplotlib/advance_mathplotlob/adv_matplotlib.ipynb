# Annotation of Graph
it is used to add text on the graph just to name a point 

```python
batters = pd.read_csv('batter.csv')
```

we get top 100 batsman in IPL 2017 and then extract the random 25 player from it 
and plot them as a scatter plot

```python
sample_df =batters.head(100).sample(25,random_state = 42)
sample_df
```

```python
plt.figure(figsize=(20,18))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],color='red',fontsize=15,alpha=.8)
```

# Horizontal and vertical line 


```python
plt.figure(figsize=(20,18))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='orange',linestyle = '--') # horizontal line 
plt.axvline(30,color='orange',linestyle = '--') # vertical line

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1,figsize=(20,18),sharex=True )
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('avg vs strike rate')
ax[0].set_ylabel('Strike RATE')


ax[1].set_title('avg vs Runs')
ax[1].set_ylabel('RUNS')
ax[1].set_xlabel('AVERAGE ')
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2,figsize=(20,18) )
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])

ax[1,1].hist(batters['runs'])
```

# 3d surfaces  

```python
# x^2 + y^2 

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
# step 2 create a meshgrid 
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2 
print(z.shape)
```

```python
fir = plt.figure(figsize=(13,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap  = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)+np.cos(yy)
fig = plt.figure(figsize=(13,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap  = 'viridis')
fig.colorbar(p)
```

# Contour Graph 
3D --> 2D Graph

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap  = 'viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap  = 'viridis')
fig.colorbar(p)
```
